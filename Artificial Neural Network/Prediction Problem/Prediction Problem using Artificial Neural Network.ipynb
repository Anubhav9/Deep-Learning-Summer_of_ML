```python
data_x=data_x.values
```

```python
data_y=data_y.values
```

```python
data_y=data_y.reshape((2250,))
```

```python
data_y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(data_x)
sc.transform(data_y.reshape(-1,1))
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(output_dim=2,init="uniform",activation="relu",input_dim=2))
```

```python
classifier.add(Dense(output_dim=1,init="uniform",activation="sigmoid"))
```

```python
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
classifier.fit(data_x,data_y,batch_size=32,nb_epoch=10)
```

```python
data_test=pd.read_csv('C:/Users/Anubhav Sanyal/Desktop/Deep Learning Summer of ML-CodingBlocks/ANN/Logistic_X_Test (1).csv')
```

```python
t=classifier.predict(data_test)
```
