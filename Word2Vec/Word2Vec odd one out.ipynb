```python
from gensim.models import KeyedVectors 
```

```python
from gensim.models import Word2Vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectors=KeyedVectors.load_word2vec_format('C:/Coding Blocks-Assignments/Word2vec Assignment/GoogleNews-vectors-negative300.bin',binary=True)
```

```python
import numpy as np
```

```python
def out(words):
    allvectors=[vectors[w] for w in words]
    #print(allvectors)
    average=np.mean(allvectors,axis=0)
    min=1.0
    odd=None
    for w in words:
        similarity=cosine_similarity([average],[vectors[w]])
        if(similarity<min):
            min=similarity
            odd=w
    return odd
        
```

```python
list=['bat','Chemistry','Physics']
```

```python
k=out(list)
```
